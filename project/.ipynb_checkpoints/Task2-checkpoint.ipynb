```python
def normalize_column_names(columns):
    return [column.lower() for column in columns]
    

def add_index(df):
    df.index.rename('id', inplace=True)
    df.index += 1
    return 
```

### chem_gene

```python
columns_cg_str = "ChemicalName,ChemicalID,CasRN,GeneSymbol,GeneID,GeneForms,Organism,OrganismID,Interaction,InteractionActions,PubMedIDs"
columns_cg = columns_cg_str.split(",")
columns_final_cg = normalize_column_names(columns_cg)

column_cg_dtype = {"chemicalname": str,
                "chemicalid": str,
                "casrn": str,
                "genesymbol": str,
                "geneid": str,
                "geneforms": str,
                "organism": str,
                "organismid": str,
                "interaction": str,
                "interactionactions": str,
                "pubmedids": str}
```

```python
df_chem_gene = pd.read_csv("http://ctdbase.org/reports/CTD_chem_gene_ixns.tsv.gz", compression='gzip', header=0, sep='\t', comment='#', names=columns_final_cg, low_memory=False, dtype=column_cg_dtype)
df_chem_gene.head(3)
```

#### Drop rows with missing CasRN 

```python
df_chem_gene['casrn'].isnull().sum()
```

```python
df_chem_gene.dropna(subset='casrn', inplace=True)
```

#### Drop all rows where organism id is not 9606 (Homo sapiens)

```python
df_chem_gene.drop(df_chem_gene[df_chem_gene['organismid'] != '9606'].index, inplace = True)
df_chem_gene.head()
```

#### Split and explode InteractionActions column

```python
df_chem_gene['intact'] = df_chem_gene.interactionactions.str.split('|')
```

```python
df_chemical_gene = df_chem_gene.explode('intact')
df_chemical_gene.drop(labels=['interactionactions', 'geneforms', 'pubmedids'], axis=1, inplace=True)
df_chemical_gene.rename(columns={'intact': 'interactionactions'}, inplace=True)
df_chemical_gene.head(2)
#df_chemical_gene.rename(columns={'IntAct': 'InteractionActions'}, inplace=True)
```

#### Add a new index

```python
df_chemical_gene.reset_index(inplace=True)
df_chemical_gene.index.rename('id', inplace=True)
df_chemical_gene.drop('index', axis=1, inplace=True)
df_chemical_gene.index += 1
```

#### Change datatypes for columns organismid and geneid

```python
df_chemical_gene[['geneid', 'organismid']] = df_chemical_gene[['geneid', 'organismid']].apply(pd.to_numeric)
```

## Create SQLAlchemy model

```python
"""UniProt RDBMS model definition."""
from sqlalchemy.orm import relationship
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import Column, Integer, String, Table, Text, ForeignKey, DateTime, Float

from collections import defaultdict

Base = declarative_base()

class ChemicalGene(Base):
    """Class definition for the Compounds table."""
    __tablename__ = 'chemical_gene_interactions'

    id = Column(Integer, primary_key=True)
    chemicalname = Column(String(255), nullable=False)
    chemicalid = Column(String(32), nullable=False)
    casrn = Column(String(32), nullable=False)
    genesymbol = Column(String(32), nullable=False)
    geneid = Column(Integer, nullable=False)
    organism = Column(String(32), nullable=False)
    organismid = Column(Integer, nullable=False)
    interaction = Column(String(1024), nullable=False)
    interactionactions = Column(String(255), index=True, nullable=False)
    
    
class ChemicalDisease(Base):
    """Class definition for the ChemicalData table."""
    __tablename__ = 'chemical_disease_associations'

    id = Column(Integer, primary_key=True)
    chemicalid = Column(String(32), nullable=False)
    diseasename = Column(String(1024), nullable=False)
    diseaseid = Column(String(32), nullable=False)
    inferencescore = Column(Float, nullable=True)
```

```python
def filter_chemical_disease_association(chemical_ids: List, disease:str = "PARKINSON") -> List:
    """Filters out chemical ids that do not correspond to Parkinson's 
        disease from the chmeical_disease_associations table
        Returns
        -------
        filtered_chemical_ids: List
            List of chemical ids that are associated with Parkinson's disease.            
    """ 
    chem_ids = []
    entries = session.query(ChemicalDisease).filter(ChemicalDisease.diseasename.like(f"%{disease}%")).all()
    for entry in entries:
        if entry.chemicalid in chemical_ids:
            chem_ids.append(entry.chemicalid)
    return list(set(chem_ids))
```
